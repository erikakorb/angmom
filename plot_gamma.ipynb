```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

## Gamma as function of q

```python
# gamma factor as function of mass ratio
# note: this is gamma as used in SEVN and Sobermann+1997
def gamma_Soberman(q, rlo_gamma_angmom):
    if rlo_gamma_angmom == '-1':   # Jeans
        gamma = (q/(1.0+q))**2
    elif rlo_gamma_angmom == '-2':
        gamma = (1/(1.0+q))**2
    return gamma

def gamma_Pols(q, gamma_Soberman):
    factor = (1.0+q)**2 / q
    return factor * gamma_Soberman




plt.rcParams.update({
"text.usetex": True,
"text.latex.preamble": r"\boldmath",
"font.family": "cmr10",   # computer modern font, as in latex
"axes.formatter.use_mathtext": True,
"font.size": 20,
"font.weight": "bold",
"axes.labelsize" : 20,
"xtick.labelsize" : 20,
"ytick.labelsize" : 20,
"legend.fontsize": 17
})



q = np.linspace(0,1.5, 50)
gamma_Jeans_Sob = gamma_Soberman(q, rlo_gamma_angmom= '-1')
gamma_iso_Sob = gamma_Soberman(q, rlo_gamma_angmom='-2')
gamma_Jeans_Pols = gamma_Pols(q, gamma_Jeans_Sob)    # equivalent to q
gamma_iso_Pols = gamma_Pols(q, gamma_iso_Sob)        # equivalent to 1/q


fig = plt.figure(figsize=(7.1,6))
plt.plot(q,gamma_Jeans_Sob, linestyle='dashed', color='r',  lw=3, label=r'\textbf{Jeans mode}')
plt.plot(q,gamma_iso_Sob, linestyle='solid', color='k', lw=3,  label=r'\textbf{Isotropic re-emission}')
plt.axhline(y=1, linestyle='dotted',color='k', lw=1,zorder=1)

plt.xlim(xmin=-0.02,xmax=1.52)
plt.ylim(ymin=-0.02, ymax=1.02)
plt.legend(loc='upper right', ncols=1)
#plt.yscale('log')
plt.xlabel(r'$q=M_{\rm acc}/M_{\rm don}$')
plt.ylabel(r'$\gamma_{\rm Soberman} = \gamma_{\rm SEVN}$')
plt.tight_layout(rect=(-0.03, -0.03, 1.02, 1.03))
plt.show()
os.makedirs('./plots_gamma', exist_ok=True)
fig.savefig('./plots_gamma/Soberman.pdf', dpi=300)


fig = plt.figure(figsize=(7.1,6))
plt.plot(q,gamma_Jeans_Pols, linestyle='dashed', color='r',  lw=3, label=r'\textbf{Jeans mode}')
plt.plot(q,gamma_iso_Pols, linestyle='solid', color='k', lw=3,  label=r'\textbf{Isotropic re-emission}')
plt.axhline(y=1, linestyle='dotted',color='k', lw=1,zorder=1)

plt.xlim(xmin=-0.02,xmax=1.52)
plt.ylim(ymin=-0.02, ymax=3.02)
plt.legend(loc='upper right', ncols=1)
#plt.yscale('log')
plt.xlabel(r'$q=M_{\rm acc}/M_{\rm don}$')
plt.ylabel(r'$\gamma_{\rm Pols}$')
plt.tight_layout(rect=(-0.03, -0.03, 1.02, 1.04))
plt.show()
os.makedirs('./plots_gamma',exist_ok=True)
fig.savefig('./plots_gamma/Pols.pdf', dpi=300)
```
